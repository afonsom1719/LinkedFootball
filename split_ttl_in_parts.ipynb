```python
# File splitting script
input_file = "sports_data.ttl"  # Path to your input file
output_prefix = "sports_data_part"  # Prefix for output files
max_chunks = 100  # Maximum number of chunks per file

def split_turtle_file(input_file, output_prefix, max_chunks):
    with open(input_file, "r", encoding="utf-8") as infile:
        lines = infile.readlines()

    # Extract the prefix lines
    prefix_lines = []
    for line in lines:
        if line.startswith("@prefix"):
            prefix_lines.append(line)
        else:
            break

    # Extract the data chunks
    chunks = []
    current_chunk = []
    for line in lines[len(prefix_lines):]:
        current_chunk.append(line)
        if line.strip().endswith("."):  # A chunk ends when a line ends with "."
            chunks.append(current_chunk)
            current_chunk = []

    if current_chunk:  # Add any remaining chunk
        chunks.append(current_chunk)

    # Split the chunks into multiple files
    num_files = (len(chunks) + max_chunks - 1) // max_chunks
    for i in range(num_files):
        start = i * max_chunks
        end = start + max_chunks
        chunk_subset = chunks[start:end]

        # Write to a new file
        output_file = f"{output_prefix}_{i + 1}.ttl"
        with open(output_file, "w", encoding="utf-8") as outfile:
            # Write the prefix lines
            outfile.writelines(prefix_lines)
            # Write the chunks
            for chunk in chunk_subset:
                outfile.writelines(chunk)

        print(f"Created file: {output_file}")

# Run the function
split_turtle_file(input_file, output_prefix, max_chunks)
```
