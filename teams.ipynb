```python
import requests
import csv
from rapidfuzz import process

competitions_ids = ['PL', 'PD', 'SA', 'BL1', 'PPL']
csv_file_list = ['GB1_club_values.csv', 'ES1_club_values.csv', 'IT1_club_values.csv', 'L1_club_values.csv', 'PO1_club_values.csv']
csv_file_list_players = ['GB1_players_data.csv', 'ES1_players_data.csv', 'IT1_players_data.csv', 'L1_players_data.csv', 'PO1_players_data.csv']
csv_file_list = [f'tm_values/{csv_file}' for csv_file in csv_file_list]
csv_file_list_players = [f'tm_values/{csv_file}' for csv_file in csv_file_list_players]

for competition_id, csv_file, csv_file_players in zip(competitions_ids, csv_file_list, csv_file_list_players):
        
            url = f'http://api.football-data.org/v4/competitions/{competition_id}/teams'
            headers = {'X-Auth-Token': '644defeac7fd49b4a15c87fbc6e6f6e5'}
            response = requests.get(url, headers=headers)
            data = response.json()

            with open(csv_file, mode='r') as infile:
                reader = csv.reader(infile)
                next(reader)  # Skip the header row
                mydict = {rows[0]:rows[1] for rows in reader}

            with open(csv_file_players, mode='r') as infile:
                reader = csv.reader(infile)
                next(reader)  # Skip the header row
                mydict_players = {rows[0]: (rows[2], rows[3]) for rows in reader if rows[0]}  # Filter out empty player names
                print(mydict_players)

            with open(f'data/competition_players_{competition_id}.csv', mode='w') as outfile:
                writer = csv.writer(outfile)
                writer.writerow(['Id', 'Name', 'Position', 'Date of Birth', 'Country', 'Market Value', 'Team', 'TeamId', 'PictureUrl'])

                for team in data['teams']:

                    name = team['name']
                    tid = team['id']
                    for player in team['squad']:
                        id = player['id']
                        player_name = player['name']
                        position = player['position']
                        date_of_birth = player['dateOfBirth']
                        country = player['nationality']

                        # get value by similarity
                        result = process.extractOne(player_name.lower(), mydict_players.keys())
                        if result:
                            match, score, _ = result  
                            if score > 50:  
                                market_value, picture_url = mydict_players[match]
                            else:
                                print("No suitable match found.")
                                market_value = "0"
                                picture_url = ""
                        else:
                            print("No matches returned.")
                            market_value = "0"
                            picture_url = ""

                        writer.writerow([id, player_name, position, date_of_birth, country, market_value, name, tid, picture_url])
                        print(id, player_name, position, date_of_birth, country, market_value, name, tid, picture_url)
```
